```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load dataset
df = pd.read_csv("diabetes.csv")

# Features (excluding target)
X = df.drop("Outcome", axis=1)
X_train, _, = train_test_split(X, test_size=0.2, random_state=0)

# Normalize features
X_norm = normalize(X_train)

# KMeans clustering (k=3)
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
kmeans.fit(X_norm)

# Silhouette Score
print("Silhouette Score (k=3):", silhouette_score(X_norm, kmeans.labels_))

# Cluster visualisation with two features
sns.scatterplot(x=X_train["Glucose"], y=X_train["BMI"], hue=kmeans.labels_, palette="Set2")
plt.title("KMeans Clusters (based on Glucose vs BMI)")
plt.show()

# Silhouette scores for different k values
K = range(2, 7)
scores = [silhouette_score(X_norm, KMeans(n_clusters=k, random_state=0, n_init=10).fit_predict(X_norm)) for k in K]

# Line plot for silhouette scores
sns.lineplot(x=list(K), y=scores, marker='o')
plt.title("Silhouette Score vs Number of Clusters")
plt.xlabel("k (Number of clusters)")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score

# Load dataset
df = pd.read_csv("diabetes.csv")

# Features only (excluding target)
X = df.drop("Outcome", axis=1)
X_train, _, = train_test_split(X, test_size=0.2, random_state=0)

# Normalize
X_norm = normalize(X_train)

# Fit KMeans (k=3)
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
kmeans.fit(X_norm)

# Silhouette score
print("Silhouette Score (k=3):", silhouette_score(X_norm, kmeans.labels_))

# Cluster visualization using two features
sns.scatterplot(x=X_train["Glucose"], y=X_train["BMI"], hue=kmeans.labels_, palette="Set2")
plt.title("Clusters by Glucose vs BMI")
plt.show()

# Try different values of k
K = range(2, 7)
scores = [silhouette_score(X_norm, KMeans(n_clusters=k, random_state=0, n_init=10).fit(X_norm).labels_) for k in K]

# Plot Silhouette Score vs K
sns.lineplot(x=list(K), y=scores, marker='o')
plt.title("Silhouette Score vs Number of Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()
```
