```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics

# Load dataset
df = pd.read_csv('Salary_Data.csv')
X = df.iloc[:, :1]
y = df.iloc[:, 1:]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Plotting function
def plot_results(X, y, y_pred, title):
    plt.scatter(X, y, color='coral')
    plt.plot(X, y_pred, color='firebrick')
    plt.title(title)
    plt.xlabel('Years of Experience')
    plt.ylabel('Salary')
    plt.show()

# Visualize results
plot_results(X_train, y_train, y_pred_train, 'Training Set')
plot_results(X_test, y_test, y_pred_test, 'Test Set')


# Print model coefficients
print(f'Coefficient: {model.coef_[0][0]:.2f}')
print(f'Intercept: {model.intercept_[0]:.2f}')

# Evaluation metrics
print(f'MAE: {metrics.mean_absolute_error(y_test, y_pred_test):.2f}')
print(f'MSE: {metrics.mean_squared_error(y_test, y_pred_test):.2f}')
print(f'RMSE: {np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)):.2f}')
```
