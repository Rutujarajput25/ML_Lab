```python
# Loading the dataset from a CSV file into a pandas DataFrame
data = pd.read_csv("used_cars.csv")  
```

```python
# Remove S.No. column from data
data = data.drop(['S.No.'], axis = 1)
data.info()
```

```python
from datetime import date  # Importing date module to get the current year  

# Getting the current year  
current_year = date.today().year  

# Calculating the age of the car by subtracting the manufacturing year from the current year  
data['Car_Age'] = current_year - data['Year']  

# Displaying the first five rows of the dataset to verify the new 'Car_Age' column  
data.head()
```

```python
# Extracting the brand name of the car from the 'Name' column (assuming the first word is the brand)  
data['Brand'] = data.Name.str.split().str.get(0)
```

```python
# Extracting the car model from the 'Name' column by combining the second and third words  
data['Model'] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2)
```

```python
# Displaying the 'Name', 'Brand', and 'Model' columns to check the extracted values  
data[['Name', 'Brand', 'Model']]
```

```python
# Displaying the unique car brands in the dataset  
print(data.Brand.unique())  

# Displaying the number of unique car brands in the dataset  
print(data.Brand.nunique())
```

```python
# Searching for rows where the 'Brand' column contains any of the specified brands (case-sensitive search)  
searchfor = ['Isuzu', 'ISUZU', 'Mini', 'Land']  # List of brands to search for
data[data.Brand.str.contains('|'.join(searchfor))].head(5)  # Displaying the first 5 rows of the filtered data
```

```python
# Replacing specific brand names with corrected or standardized names in the 'Brand' column
data["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper", "Land": "Land Rover"}, inplace=True)
```

```python
# Generating descriptive statistics for the numerical columns and transposing the result for better readability  
data.describe().T

```

```python
# Generating descriptive statistics for all columns (including categorical ones) and transposing the result for better readability  
data.describe(include='all').T
```

```python
# Identifying categorical columns (columns with object data type)
cat_cols = data.select_dtypes(include=['object']).columns  

# Identifying numerical columns (columns with numerical data types)
num_cols = data.select_dtypes(include=np.number).columns.tolist()

# Printing the list of categorical variables
print("Categorical Variables:")
print(cat_cols)

# Printing the list of numerical variables
print("Numerical Variables:")
print(num_cols)
```

```python
# Looping through each numerical column to calculate its skewness and visualize its distribution and boxplot
for col in num_cols:
    print(col)  # Printing the column name
    print('Skew :', round(data[col].skew(), 2))  # Printing the skewness of the column (rounded to 2 decimal places)

    # Creating a figure with two subplots to visualize the column
    plt.figure(figsize = (15, 4))  
    
    # Histogram plot to visualize the distribution of the column
    plt.subplot(1, 2, 1)  
    data[col].hist(grid=False)  
    plt.ylabel('count')  # Labeling the y-axis as 'count'
    
    # Boxplot to visualize the spread and identify any outliers
    plt.subplot(1, 2, 2)  
    sns.boxplot(x=data[col])  
    
    plt.show()  # Display the plots
```

```python
# Creating a 3x2 grid of subplots to display bar plots for various categorical variables
fig, axes = plt.subplots(3, 2, figsize=(18, 18))

# Adding a title for the entire figure
fig.suptitle('Bar plot for all categorical variables in the dataset')

# Creating count plots for each categorical column
sns.countplot(ax=axes[0, 0], x='Fuel_Type', data=data, color='blue', 
              order=data['Fuel_Type'].value_counts().index)  # Fuel_Type distribution
sns.countplot(ax=axes[0, 1], x='Transmission', data=data, color='blue', 
              order=data['Transmission'].value_counts().index)  # Transmission distribution
sns.countplot(ax=axes[1, 0], x='Owner_Type', data=data, color='blue', 
              order=data['Owner_Type'].value_counts().index)  # Owner_Type distribution
sns.countplot(ax=axes[1, 1], x='Location', data=data, color='blue', 
              order=data['Location'].value_counts().index)  # Location distribution
sns.countplot(ax=axes[2, 0], x='Brand', data=data, color='blue', 
              order=data['Brand'].head(20).value_counts().index)  # Brand distribution (top 20 brands)
sns.countplot(ax=axes[2, 1], x='Model', data=data, color='blue', 
              order=data['Model'].head(20).value_counts().index)  # Model distribution (top 20 models)

# Rotating x-axis labels for better readability
axes[1][1].tick_params(labelrotation=45)  # Rotate Location labels
axes[2][0].tick_params(labelrotation=90)  # Rotate Brand labels
axes[2][1].tick_params(labelrotation=90)  # Rotate Model labels

# Display the plots
plt.show()
```

```python
# Function for log transformation of the column
def log_transform(data, cols):
    for colname in cols:
        # Checking if all values in the column are equal to 1
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname] + 1)  # Apply log transformation with +1 for the '1.0' values
        else:
            data[colname + '_log'] = np.log(data[colname])  # Apply log transformation to the column

# Applying the log transformation to the specified columns
log_transform(data, ['Kilometers_Driven', 'Price'])

# Check the updated info of the dataset
data.info()
```

```python
fig, axarr = plt.subplots(4, 2, figsize=(12, 18))
data.groupby('Location')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12)
axarr[0][0].set_title("Location Vs Price", fontsize=18)
data.groupby('Transmission')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12)
axarr[0][1].set_title("Transmission Vs Price", fontsize=18)
data.groupby('Fuel_Type')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12)
axarr[1][0].set_title("Fuel_Type Vs Price", fontsize=18)
data.groupby('Owner_Type')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12)
axarr[1][1].set_title("Owner_Type Vs Price", fontsize=18)
data.groupby('Brand')['Price_log'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][0], fontsize=12)
axarr[2][0].set_title("Brand Vs Price", fontsize=18)
data.groupby('Model')['Price_log'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][1], fontsize=12)
axarr[2][1].set_title("Model Vs Price", fontsize=18)
data.groupby('Seats')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=12)
axarr[3][0].set_title("Seats Vs Price", fontsize=18)
data.groupby('Car_Age')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=12)
axarr[3][1].set_title("Car_Age Vs Price", fontsize=18)
plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
sns.despine()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Drop unwanted columns, ignoring errors if they are missing
data_filtered = data.drop(columns=['Kilometers_Driven', 'Price'], errors='ignore')

# Select only numeric columns
data_numeric = data_filtered.select_dtypes(include=['number'])

# Plot the heatmap
plt.figure(figsize=(12, 7))
sns.heatmap(data_numeric.corr(), annot=True, vmin=-1, vmax=1, cmap="coolwarm")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Copy the data to avoid modifying the original DataFrame
data_encoded = data.copy()

# Identify categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns

# Apply Label Encoding (converts categories to numbers)
for col in categorical_cols:
    data_encoded[col] = LabelEncoder().fit_transform(data_encoded[col])

# Drop unwanted columns
data_encoded = data_encoded.drop(columns=['Kilometers_Driven', 'Price'], errors='ignore')

# Compute correlation
plt.figure(figsize=(12, 7))
sns.heatmap(data_encoded.corr(), annot=True, vmin=-1, vmax=1, cmap="coolwarm")
plt.show()
```
